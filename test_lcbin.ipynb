```python
import lcbin as lcb
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate some fake data
t = np.linspace(0, 1, 20000)
f = np.ones(len(t))
x = np.argwhere(abs(t - 0.5) < 0.05)
f[x] -= 0.0005
scatter = 0.001
e = np.ones(len(t)) * scatter
f = f + np.random.normal(0.0, scatter, len(t))
plt.figure(figsize = (20, 14))
plt.plot(t,f,'.', alpha = 0.1)

# Now, you can bin the light curve to get a smaller number of more precise points
tbin, fbin, ebin, mask = lcb.lcbin(t,f, 100) 
plt.errorbar(tbin, fbin, yerr = ebin, fmt = 'o', color = 'red')
```

```python
#by default, lcbin estimates the uncertainty in each bin, but if you know the uncertainty, you can get better
#estimates of the uncertainty on the bin. 

plt.figure(figsize = (20, 14))
plt.plot(t,f,'.', alpha = 0.1)

tbin, fbin, ebin, mask = lcb.lcbin(t,f, 100, errorin = e) 
plt.errorbar(tbin, fbin, yerr = ebin, fmt = 'o', color = 'red')
```
